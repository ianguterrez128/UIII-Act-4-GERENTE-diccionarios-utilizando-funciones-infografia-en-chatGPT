22. Gerente: Nivel de autoridad, unidad de negocio, bonos por desempeño, estrategia principal, reuniones diarias y sucursal.

```python
# ======================================================
# SISTEMA DE GERENTES
# CBTIS 128
# Especialidad: Inteligencia Artificial
# Submodulo 2: Elabora proyectos de programación lógica
# Docente: Eliseo Nava
# Alumno: Ian Josafat Gutierrez Saucedo
# ======================================================

# diccionario vacio
gerentes = []


# ======================================================
# FUNCION AGREGAR
# ======================================================

def agregar_gerente():

    while True:

        nombre = input("ingresa el nombre del gerente: ")
        autoridad = input("ingresa el nivel de autoridad: ")
        unidad = input("ingresa la unidad de negocio: ")
        bonos = input("ingresa bonos por desempeño: ")
        estrategia = input("ingresa la estrategia principal: ")
        reuniones = input("ingresa reuniones diarias: ")
        sucursal = input("ingresa la sucursal: ")

        gerente = {
            "nombre": nombre,
            "autoridad": autoridad,
            "unidad": unidad,
            "bonos": bonos,
            "estrategia": estrategia,
            "reuniones": reuniones,
            "sucursal": sucursal
        }

        gerentes.append(gerente)

        print("\ngerente agregado correctamente")

        respuesta = input("deseas agregar otro gerente? si/no: ")

        if respuesta == "no":
            break


# ======================================================
# FUNCION LISTAR
# ======================================================

def listar_gerentes():

    while True:

        if len(gerentes) == 0:

            print("\nno hay gerentes registrados")

        else:

            print("\n================ TABLA DE GERENTES ================")

            print("--------------------------------------------------------------------------")
            print("NOMBRE\tAUTORIDAD\tUNIDAD\tBONOS\tESTRATEGIA\tREUNIONES\tSUCURSAL")
            print("--------------------------------------------------------------------------")

            for gerente in gerentes:

                for key, value in gerente.items():
                    print(value, end="\t")

                print()

            print("--------------------------------------------------------------------------")

        respuesta = input("\ndeseas volver a listar? si/no: ")

        if respuesta == "no":
            break


# ======================================================
# FUNCION BUSCAR
# ======================================================

def buscar_gerente():

    while True:

        buscar = input("ingresa el nombre del gerente a buscar: ")

        encontrado = False

        for gerente in gerentes:

            if gerente["nombre"] == buscar:

                print("\n=========== GERENTE ENCONTRADO ===========")

                print("------------------------------------------")

                for key, value in gerente.items():

                    print(key, ":", value)

                print("------------------------------------------")

                encontrado = True

        if encontrado == False:

            print("\ngerente no encontrado")

        respuesta = input("\ndeseas buscar otro gerente? si/no: ")

        if respuesta == "no":
            break


# ======================================================
# FUNCION BORRAR
# ======================================================

def borrar_gerente():

    while True:

        borrar = input("ingresa el nombre del gerente a borrar: ")

        encontrado = False

        for gerente in gerentes:

            if gerente["nombre"] == borrar:

                gerentes.remove(gerente)

                print("\ngerente eliminado correctamente")

                encontrado = True

                break

        if encontrado == False:

            print("\ngerente no encontrado")

        respuesta = input("\ndeseas borrar otro gerente? si/no: ")

        if respuesta == "no":
            break


# ======================================================
# FUNCION ACTUALIZAR
# ======================================================

def actualizar_gerente():

    while True:

        buscar = input("ingresa el nombre del gerente a actualizar: ")

        encontrado = False

        for gerente in gerentes:

            if gerente["nombre"] == buscar:

                gerente["nombre"] = input("nuevo nombre: ")
                gerente["autoridad"] = input("nuevo nivel de autoridad: ")
                gerente["unidad"] = input("nueva unidad de negocio: ")
                gerente["bonos"] = input("nuevos bonos por desempeño: ")
                gerente["estrategia"] = input("nueva estrategia principal: ")
                gerente["reuniones"] = input("nuevas reuniones diarias: ")
                gerente["sucursal"] = input("nueva sucursal: ")

                print("\ngerente actualizado correctamente")

                encontrado = True

        if encontrado == False:

            print("\ngerente no encontrado")

        respuesta = input("\ndeseas actualizar otro gerente? si/no: ")

        if respuesta == "no":
            break


# ======================================================
# MENU PRINCIPAL
# ======================================================

while True:

    print("\n================ MENU DE GERENTES ================")
    print("1. agregar gerente en un diccionario")
    print("2. listar diccionarios de gerentes")
    print("3. buscar gerente en diccionario")
    print("4. borrar gerente en diccionario")
    print("5. actualizar gerente en diccionario")
    print("6. salir")

    opcion = int(input("\ningresa una opcion: "))

    if opcion == 1:

        agregar_gerente()

    elif opcion == 2:

        listar_gerentes()

    elif opcion == 3:

        buscar_gerente()

    elif opcion == 4:

        borrar_gerente()

    elif opcion == 5:

        actualizar_gerente()

    elif opcion == 6:

        print("\nfin del programa")
        break

    else:

        print("\nopcion no valida")
```
